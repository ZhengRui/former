### Basic Concepts:
* #### GIS Data Type:
    * vector data: include points, lines and polygons.<br/><br/>
    <img src='files/images/VectorEarth.png' style="height: 100px;" alt='vector earth' title='vector earth'/>
    <br/><br/>
    * raster data: high resolution image.<br/><br/>
    <img src='files/images/RasterEarth.jpg' style="height: 100px;" alt='raster earth' title='raster earth'/>
    <br/><br/>
* #### GIS Data Format (Vector Format):
    * ShapeFile: contains a collection of files
        + .shp : shape format
        + .shx : index format
        + .dbf : attribute format
        + maybe other files: .prj, .sbn ...
    * OSM File: .osm file is actually xml type data <br/>
        * contains three elements:
            + node: lat, lon
            + way: node ids in the way
            + member: logical or geographic relationship between nodes and ways
        * supplementary information about each element contained in sub tag element, which are the possible features in the map. All features and their possible values with meanings are listed in this page: [OSM features](http://wiki.openstreetmap.org/wiki/Map_features). In the sub tag element, it has the form of key-value pairs, key maybe one of these features, while value maybe one of the possible values which corresponds to this feature. As i only care about road network, i just need to care about the way elements with highway keyword in its tag, the value of highway indicates this is a motorway, or trunk or a link..., other keywords like 'tunnel' or 'bridge' could give additional information.
        
* #### Gis Data Source (Vector Format):
    * Open Street Map: two ways:
        + export directly from [Open Street Map](http://www.openstreetmap.org), however, it seems with this method it can export map data only when the zoom level is higher than some value.
        + download whole planet or countries or areas from [OSM planet.osm](http://wiki.openstreetmap.org/wiki/Planet.osm#Downloading).
        + in python script, we can define such funcion to get data of small area:
        
    ```python
        def download_osm(l,b,r,t):
            """ Return a filehandle to the downloaded data."""
            from urllib import urlopen
            fp = urlopen( "http://api.openstreetmap.org/api/0.6/map?bbox=%f,%f,%f,%f"%(l,b,r,t) )
            return fp
    ```
        
    * [TIGER](http://www.census.gov/geo/www/tiger/index.html) (Topologically Integrated Geographic Encoding and Referencing System). The TIGER data includes information on streets, railways, rivers, lakes, geographic boundaries, and legal and statistical areas such as school districts, and urban regions. Seems can not open outside of US.
    * [Natural Earth]( http://www.naturalearthdata.com).
    * [GSHHS](http://www.ngdc.noaa.gov/mgg/shorelines/gshhs.html)(Global self-consistent, hierarchical, high-resolution 
shoreline database).
    * [World Boarders Dataset](http://thematicmapping.org/downloads/world_borders.php).
    
* #### Spatial Reference System Identifier (SRID):
    * Projection: Creating a two-dimensional map from the three-dimensional shape of the earth is a process known as projection.
        * Cylindrical projections:<br/><br/>
        <img src='files/images/cyproj1.png' style="height: 100px;"/>
        <img src='files/images/cyproj2.png' style="height: 100px;"/><br/><br/>
        include *Mercator Projection, Equal-Area Cylindrical Projection, and Universal Transverse Mercator Projection*.
        * Conic projcetions:<br/><br/>
        <img src='files/images/coproj1.png' style="height: 100px;"/>
        <img src='files/images/coproj2.png' style="height: 100px;"/><br/><br/>
        include *Albers Equal-Area Projection, Lambert Conformal Conic Projection, and Equidistant Projection*.
        * Azimuthal projections:<br/><br/>
        <img src='files/images/azproj1.png' style="height: 100px;"/>
        <img src='files/images/azproj2.png' style="height: 100px;"/><br/><br/>
        Azimuthal projections are centered around a single point, and don't generally show the entire earth's surface, include * Gnomonic Projection, Lambert Equal-Area Azimuthal Projection, and the Orthographic Projection*.
        * the mathematician Carl Gausse proved that it is mathematically impossible to project a three-dimensional shape such as a sphere onto a flat plane without introducing some sort of distortion. This is why there are so many different types of projections: some projections are more suited to a given purpose, but no projection can do everything.
        * **Universal Transverse Mercator (UTM)** coordinate 
system divides the world up into 60 different "zones", each zone using a different map projection to minimize projection errors. Within a given zone, the coordinates are measured as the number of meters away from the zone's origin, which is the intersection of the equator and the central meridian for that zone. False northing 
and false easting values are then added to the distance in meters away from this reference point to avoid having to deal with negative numbers.<br/><br/>
        <img src='files/images/falsene.png' style="height: 200px;"/><br/><br/>
    * Datum: Roughly speaking, a datum is a mathematical model of the earth used to shape of the earth's surface. 
        * NAD 27
        * NAD 83
        * WGS 84
        * When dealing with geospatial data covering the United States, WGS 84 is basically identical to NAD 83. WGS 84 also has the distinction of being used by Global Positioning System satellites, so all data captured by GPS units will use this datum. NAD 83 is also used very often, while NAD 27 is least frequently used.
    * SRID: is a unique value used to unambiguously identify projected, unprojected, and local spatial coordinate system definitions. For example, WGS 84 - SRID 4326, NAD 27 - SRID 2029. 
OSM and google map use SRID 900913. In postgis database, there will be a talbe called "spatial_ref_sys", it saves the definition for each srid, for example to see what is SRID 900913:

```SQL
select * from spatial_ref_sys where srid=900913
```

To see how to transfrom between different SRID in PostGis, check this: [Section 15: Projecting Data](http://workshops.boundlessgeo.com/postgis-intro/projection.html). In .osm file, the lat and lon are unprojected WSG84 coordinates, while after projected and exported as .xml file (or after imported into PostGis using Osm2pgsql tool with default option), it transforms into SRID 900913 coordinates, you will see lat and lon as a big number which use meter as unit. Obviously using projected coordinates is much faster and more convenient for rendering maps and analysis.

### Basemap examples:
* #### Make a simple earth map

````python
%pylab inline
%reset -f
from mpl_toolkits.basemap import Basemap
import matplotlib.pyplot as plt
import numpy as np

# make sure the value of resolution is a lowercase L,
# for 'low', not a numeral 1
# lat_0 and lon_0 indicates the look in angle, lat_0=22.336, lon_0=114.263
# will be in hong kong
map = Basemap(projection='ortho', lat_0=50, lon_0=-100,
              resolution='l', area_thresh=1000.0)

map.drawcoastlines(linewidth=0.25)
map.drawcountries(linewidth=0.25)
map.fillcontinents(color='coral', lake_color='aqua')
map.drawmapboundary(fill_color='aqua') # fill ocean color

map.drawmeridians(np.arange(0, 360, 30))
map.drawparallels(np.arange(-90, 90, 30))

plt.show()
````

* #### Try a different projection

````python
%reset -f

from mpl_toolkits.basemap import Basemap
import matplotlib.pyplot as plt
import numpy as np
 
# make sure the value of resolution is a lowercase L,
#  for 'low', not a numeral 1
map = Basemap(projection='robin', lat_0=0, lon_0=0,
              resolution='l', area_thresh=1000.0)
 
map.drawcoastlines()
map.drawcountries()
map.fillcontinents(color='coral')
map.drawmapboundary()
 
map.drawmeridians(np.arange(0, 360, 30))
map.drawparallels(np.arange(-90, 90, 30))
 
plt.show()
````

* #### ZoomIn Using 'Merc' Projection with High Resolution

````python
%pylab inline
%reset -f
###highlight=[9]
from mpl_toolkits.basemap import Basemap
import matplotlib.pyplot as plt
import numpy as np
 
# make sure the value of resolution is a lowercase L,
#  for 'low', not a numeral 1
map = Basemap(projection='merc', lat_0=22.336, lon_0=114.263,
    resolution = 'h', area_thresh = 0.1,
    llcrnrlon=114, llcrnrlat=22,
    urcrnrlon=115, urcrnrlat=23)
 
map.drawcoastlines()
map.drawcountries()
map.fillcontinents(color='coral')
map.drawmapboundary()
 
map.drawmeridians(np.arange(0, 360, 30))
map.drawparallels(np.arange(-90, 90, 30))

plt.show()
````

* #### Plot Points

````python
###highlight=[16,17,18,19]
from mpl_toolkits.basemap import Basemap
import matplotlib.pyplot as plt
import numpy as np
 
map = Basemap(projection='merc', lat_0=22.336, lon_0=114.263,
    resolution = 'h', area_thresh = 0.1,
    llcrnrlon=114, llcrnrlat=22,
    urcrnrlon=115, urcrnrlat=23)
 
map.drawcoastlines()
map.drawcountries()
#map.fillcontinents(color = 'coral')
map.bluemarble()
map.drawmapboundary()
 
lon = 114.26286 # can also be a list
lat = 22.337617 # can also be a list
x,y = map(lon, lat) # x,y are the projected coordinates
map.plot(x, y, 'bo', markersize=5)

label = 'HKUST' # can also be a list
plt.text(x+1000, y+1000, label, color='red')
 
plt.show()
````

### PostGis + (Mapnik, TileMill, QGis) 

* Obviously the map in [OSM official](http://www.openstreetmap.org/) is not made with basemap, you may wonder how those beautiful maps which we can zoom in and zoom out were made?
    1. Render a static map with fancy styles, use command tool like *Mapnik*, or desktop tools like *TileMill*, *QGis2.0*.
    2. We need data to render the map, can directly read data from osm files, or import data in osm files into *PostGis* database, and get the data by querying this database when rendering map.
    3. To realize zoom in and zoom out, you need to understand concepts of *Map Tiles*, basically you need to set up a tile server locally, with tools like *TileStache*.
    
* Why i prefer to use mapnik and postgis?
    1. It's better to use command based tool to render a map, for the sake of later gps data overlapping and analysing.
    2. Postgis support many high level queries like 'what is the area of this park?', 'what is the nearest hotel?', which would be very cumbersome if implemented by myself.

* Installation: mainly about postgis installation and how to create database and import osm data into the created database.
    1. To install postgis, you need to first install postgresql. My version: PostgreSQL 9.1.11, Postgis 2.0.3.
    2. Create user and databases:
    
    ```bash
        sudo -u postgres createuser username
        sudo -u postgres createdb --encoding=UTF8 --owner=username gis # create a database gis owned by previous created user
        sudo -u postgres createlang plpgsql gis # note: this is obsolete with >= postgres-9.1 
    ```
    
    When creating username, it's recommanded not to add password, otherwise in later steps, to connect postgis, you need to specify password in many places of the configuration files(like stylesheet file for Mapnik, configuration file for TileStache, layers for TileMill...) of many map related tools.
    3. Activate postgis (Under my Ubuntu12.04):
    
    ```bash
        sudo -u postgres psql -d gis -f /usr/share/postgresql/9.1/contrib/postgis-2.0/postgis.sql
        sudo -u postgres psql -d gis -f /usr/share/postgresql/9.1/contrib/postgis-2.0/spatial_ref_sys.sql
        sudo -u postgres psql -d gis -c "GRANT SELECT ON spatial_ref_sys TO PUBLIC;"
        sudo -u postgres psql -d gis -c "GRANT ALL ON geometry_columns TO username;"
    ```
    
    **Note**: postgres is a default root user after install postgresql. The previous steps will generate many postgis functions in database gis, postgis2 is recommended over postgis1.5, otherwise in the late import step you will get some problems, maybe errors, which can be solved by repeating previous steps on other files start with 'legacy' in the postgis directory. After all these steps, you can enter postgresql by
    
    ```bash
        psql -U username -d gis
    ```
    
    After login postgresql, `\d` should show all the tables which already created, at least table 'spatial_ref_sys' is already created, maybe two views 'geography_columns' and 'geometry_columns' are also created by then. `\q` will quit from postgresql.
    
    This is [**A good introduction to postgis**](http://workshops.boundlessgeo.com/postgis-intro/), **notice** what are metadata tables 'spatial_ref_sys', 'geometry_columns', and 'geography_columns' about in **Section 8** and why (not) use *geography* coordinate in **Section 17**.
    4. Import osm data to postgis database: 
        + Install tools *osm2pgsql*: for ubuntu, you can easily install it after add ppa source, check [OSM-Wiki-osm2pgsql](http://wiki.openstreetmap.org/wiki/Osm2pgsql).
        + Import to postgis:
        
        ```bash
            osm2pgsql -s -U username -d gis /path/to/osm/file.osm 
        ```
        
        **Note**: it make takes a long time. '-s' means slim model, it's recommended for the sake of lowering memory consumption when importing, however it will create three intermediate tables 'planet_osm_nodes', 'planet_osm_rels' and 'planet_osm_ways'. Finally these 4 tables 'planet_osm_point', 'planet_osm_line', 'planet_osm_polygon', 'planet_osm_roads' will be generated and it is these 4 tables that will be queried by other map tools. [OSM-Wiki-osm2pgsql-schema](http://wiki.openstreetmap.org/wiki/Osm2pgsql/schema) has details about these tables, note the difference between 'planet_osm_roads' and 'planet_osm_line'.
    5. Now you can check whether everything is set up correctly by logging in postgresql and checking the genarated tables, here is my snapshot<br/><br/>
        <img src='files/images/pgcheck1.png'/><br/><br/>
        <img src='files/images/pgcheck2.png'/><br/><br/>
    *Note*: 'planet_osm_line_z\*' tables are generated by another package 'HighRoad'. It's recommended to use desktop tool **PgAdmin** to manage the postgis database. If you have met **connection refused issue**, you may have to configure postgresql's configuration file, which under my Ubuntu12.04 is '/etc/postgresql/9.1/main/pg_hba.conf', here is my configuration snapshot:<br/><br/>
    ![pg_hba.conf](files/images/pghba.png)<br/><br/>

### TileStache + (Leaflet, Openlayer) + OSMBuildings

* what is tile? [Check this page](http://www.maptiler.org/google-maps-coordinates-tile-bounds-projection/) , divide the global map into tiles based on zoom level, from 2^0 to 2^21, usually for all the map rending tools, you have to specify the stylesheet for each [OSM feature](http://wiki.openstreetmap.org/wiki/Map_Features) in different zoom levels.
* [TileStache](http://tilestache.org/doc/) is the tool (based on python) for you to generate the tiles and later render the map with web browser just by stitching tile pictures together. Unfortunately, i found rendering map locally on my machine doesn't have any advantage in speed comparing to rendering map remotely by connecting to those map servers. How to connect gps coordinate with map tile names is described in detail in [Slippy map tilenames](http://wiki.openstreetmap.org/wiki/Slippy_map_tilenames). To avoid the situation that labels are cut off near the boundary between two tiles, you need to specify the size of metatile, chech [Understanding Metatiles](https://www.mapbox.com/tilemill/docs/guides/metatiles/). Also to avoid duplicate tiles like ocean, we can use a single mbtile file to save all the tiles, chech [Storing tiles](https://www.mapbox.com/developers/mbtiles/). The mbtile file can be generated using TillMill.
* [Leaflet](http://leafletjs.com/) and [OpenLayer](http://openlayers.org/) are two js library for interactive maps.
* [OSMBuildings](http://osmbuildings.org/), when zoom in to some level, it display 3D buildings with height information.
